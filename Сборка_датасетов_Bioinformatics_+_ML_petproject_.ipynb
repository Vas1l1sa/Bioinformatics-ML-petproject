```python
#напишем функцию, которя на вход получает 2 датафрейма для сравнения, а на выход выдает списки различающихся колонок
def dif_in_dfs(data1, data2):

  columns_data1 = set(data1.columns)
  columns_data2 = set(data2.columns)

  #колонки, которые есть в data1, но нет в data2
  diff_data1 = columns_data1 - columns_data2

  #колонки, которые есть в data2, но нет в data1
  diff_data2 = columns_data2 - columns_data1

  print(f"Колонки, которые есть в одном (первом) датасете, но нет в другом (втором): {diff_data1}")
  print(f"Колонки, которые есть во втором датасете, но нет в первом: {diff_data2}")

dif_in_dfs(df1, df2)

#как мы видим, существенная часть колонок содеждит слеш. Сравним такие колонки отдельно
```

```python
#напишем функцию, которя на вход получает 2 датафрейма для сравнения, а на выход выдает списки различающихся колонок со слешем
def dif_slashes_in_dfs(data1, data2):

  #смотрим, какие колонки разные в датасетах
  columns_data1 = set(data1.columns)
  columns_data2 = set(data2.columns)

  # Колонки, которые есть в data1, но нет в data2
  diff_data1 = columns_data1 - columns_data2
  #Колонки, которые есть в df2, но нет в df1
  diff_data2 = columns_data2 - columns_data1

  #преобразуем данные из типа set в датафрейм
  columns1_list = list(diff_data1)
  diff_data1_df = pd.DataFrame(columns=columns1_list)

  columns2_list = list(diff_data2)
  diff_data2_df = pd.DataFrame(columns=columns2_list)

  #посмотрим на только на различающиеся колонки, содержащие в своем названии слеш
  diff_data1_df_with_slash = diff_data1_df.columns[diff_data1_df.columns.str.contains('/')]
  diff_data2_df_with_slash = diff_data2_df.columns[diff_data2_df.columns.str.contains('/')]
  #для удобства сравнения отсортируем колонки по алфавитному порядку
  diff_data1_df_with_slash_sorted = diff_data1_df_with_slash.sort_values()
  diff_data2_df_with_slash_sorted = diff_data2_df_with_slash.sort_values()

  print('В этих двух датафреймах различаются следующие колонки:', '\n', '\n')
  print(diff_data1_df_with_slash_sorted,'\n', '\n', '\n', diff_data2_df_with_slash_sorted)


dif_slashes_in_dfs(df1, df2)
```

**Какую информацию несут в себе колонки со слешами?**

Вы видим, что в названии каждой колонки находится более двух видов бактерий, обладающих высокой степенью генетической схожести, что затрудняет точное различие этих бактерий на уровне видов с использованием традиционных методов лабораторной диагностики, таких как культуральные методы или обычная ПЦР. Это затрудняет нам задачу по извлечению какие-то информативных данных из этих колонок, т.к. наша цель - исследовать именно отдельные виды бактерий.


Изучим различия этих видов, чтобы принять решение о том, можно ли их объединить в одну подгруппу.



Введем **критерий объединения** - схожая патогенность + среда обитания в теле:





*   Если бактерии из названий обеих сравниваемых колонок **подпадают под этот критерий**, то такие колонки можно переименовать под одинаковое название, которое бы объединило виды, записанные через слеш, в одну большую подгруппу.
*   Если бактерии из названий обеих сравниваемых колонок **не подпадают под этот критерий**, то такие колонки нужно оставить, как есть и не объединять между собой при соединении датасетов, т.к. по воздействию на тело это будут разные бактерии.

Напишем функцию для объединения всех датафреймов по общим колонкам

```python
#функция для объединения всех датафреймов по общим колонкам
def merge_dataframes(df1, df2, df3, df4):

    dataframes = [df1, df2, df3, df4]

    #начинаем объединение с первогодатафрейма
    merged_df = dataframes[0]

    #последовательно объединяем каждый датафрейм с уже объединенным
    for df in dataframes[1:]:
        #находим общие ключи
        common_columns = merged_df.columns.intersection(df.columns).tolist()
        merged_df = pd.merge(merged_df, df, how='outer', on=common_columns)
        merged_df = merged_df.astype('float64')
    return merged_df

df = merge_dataframes(df1, df2, df3, df4)
df
```

**6. Подготовка датасетов по здоровым людям и гастритникам к объединению и дальнейшему сравнению, в т.ч. методом машинного обучения**

```python
#для начала посмотрим на кол-во строк в каждом из датасетов
print('Кол-во строк в df:', len(df))
print('Кол-во строк в df_healthy:', len(df_healthy))
```

Мы видим, что в датасете со здоровыми людьми строк во много раз больше, чем в датасете с гастритниками. Поэтому уравновесим датасеты df и df_healthy по кол-ву строк для корректности дальнейшего анализа (особенно это будет важно для машинного обучения).

Будем отбирать строки из df_healthy, которые наиболее хорошо отражают среднестатистического человека из этого датасета. Для этого будем ориентироваться на квартили, т.к. такой отбор данных позволяет удалить выбросы и получить более репрезентативную выборку. Отбор строк будем производить на основании значения колонки 'not_null_sum', которая описывает среднее суммарное содержание всех видов бактерий из датасета в образцах.





```python
#посмотрим на значения not_null_sum в каждом из датасетов
print('Среднее суммарное содержание всех видов бактерий из датасета в образцах людей с гастритом:',  df['not_null_sum'].mean())
print('Среднее суммарное содержание всех видов бактерий из датасета в образцах здоровых людей:',  df_healthy['not_null_sum'].mean())
```

```python
#напишем функцию для уменьшения кол-ва строк в датасете по здоровым людям

def remove_outliers(df_healthy):
  Q1 = df_healthy['not_null_sum'].quantile(0.25)
  Q3 = df_healthy['not_null_sum'].quantile(0.75)
  IQR = Q3 - Q1

  #границы на основе IQR
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR

  #убираем выбросы
  df_healthy = df_healthy[(df_healthy['not_null_sum'] >= lower_bound) & (df_healthy['not_null_sum'] <= upper_bound)]
  return df_healthy

remove_outliers(df_healthy)

len(df_healthy)


```

Как мы видим, строк все еще очень много. Чтобы уменьшить кол-во строк и уравновесить датасеты, рандомно возьмем 140 строк из датасета, полученного после выполнения функции remove_outliers()


**Почему берем строки именно рандомным образом?**


Выбор строк рандомным образом сильно снижает временные затраты и по качеству не сильно уступает циклу по написанной выше функции remove_outliers() или новой функции, каждый раз уменьшающей порог с помощью изменяемого `factor` (lower_bound = Q1 - factor * IQR, upper_bound = Q3 + factor * IQR), т.к. основные выбросы были уже отфильтрованы функцией remove_outliers().



```python
#возьмем рандомным образом 140 сток
df_healthy = df_healthy.sample(n=140, random_state=42)
df_healthy.reset_index(drop=True, inplace=True)
df_healthy
```

**7. Создание единого большого датасета с гастритниками и здоровыми людьми**

```python
#модифицируем и применим написанную ранее функцию для объединения датафреймов
def merge_two_dataframes(df1, df2):

    dataframes = [df1, df2]
    #начинаем объединение с первогодатафрейма
    merged_df = dataframes[0]

    #последовательно объединяем каждый датафрейм с уже объединенным
    for df in dataframes[1:]:
        #находим общие ключи
        common_columns = merged_df.columns.intersection(df.columns).tolist()
        merged_df = pd.merge(merged_df, df, how='outer', on=common_columns)
        merged_df = merged_df.fillna(0)
        merged_df = merged_df.astype('float64')
    return merged_df

df_total = merge_two_dataframes(df, df_healthy)
df_total
```
